```python
# Import pustaka yang dibutuhkan
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Buat data sample
np.random.seed(1)  # Supaya hasil tetap sama

data = pd.DataFrame({
    'Umur': np.random.randint(18, 60, 1000),
    'Pendapatan': np.random.normal(5000000, 1500000, 1000).astype(int),
    'Jenis_Kelamin': np.random.choice(['Pria', 'Wanita'], 1000),
    'Status_Menikah': np.random.choice(['Menikah', 'Belum'], 1000),
    'Jumlah_Anak': np.random.randint(0, 4, 1000),
    'Pendidikan': np.random.choice(['SMA', 'D3', 'S1', 'S2'], 1000),
    'Punya_Rumah': np.random.choice([0, 1], 1000),
    'Lama_Bekerja': np.random.randint(0, 30, 1000),
    'Skor_Kredit': np.random.normal(70, 10, 1000).round(1),
    'Target_Beli': np.random.choice([0, 1], 1000)  # Target klasifikasi
})

# 2. Tampilkan data awal
print("Contoh Data:")
print(data.head())

# 3. Statistik Deskriptif
print("\nStatistik Deskriptif:")
print(data.describe())

# 4. Statistik Kategori
print("\nDistribusi Jenis Kelamin:")
print(data['Jenis_Kelamin'].value_counts())

print("\nDistribusi Status Menikah:")
print(data['Status_Menikah'].value_counts())

print("\nDistribusi Pendidikan:")
print(data['Pendidikan'].value_counts())
```
